```python
# For this question, I'll need to get the sisant database and do a left join with reta_2023_2024 on 'CODIGO_AERONAVE'

# But first, I'll need to stardize the values from ANAC in both dataframes

# Loading dataframe and Inspecting it
reta = pd.read_csv('C:\\Users\\subse\\OneDrive\\Documentos\\GitHub\\timex_consulting\\databases\\reta_2023_2024.csv')

overview(reta)
```

```python
reta = reta[['ID', 'OPERADOR', 'CNPJ', 'CADASTRO_ANAC', 'DATA_VALIDADE',
             'FABRICANTE', 'MODELO', 'NUMERO_SERIE', 'RESPONSAVEL']]

overview(reta)
```

```python
email_df = pd.merge(sisant_vencido, reta, left_on='CODIGO_AERONAVE', right_on='CADASTRO_ANAC', how='left', suffixes=('_sisant', '_reta'))

email_df = email_df.sort_values(by='RESPONSAVEL', ascending=True)

overview(email_df)

email_df.to_csv('base_drones_com_email_mais_nulos.csv')
```

```python
# Merging drones_sumiram_validacao + reta
drones_sumiram_com_email = pd.merge(drones_sumiram_validacao, reta, left_on='CODIGO_AERONAVE', right_on='CADASTRO_ANAC', how='left').sort_values(by='RESPONSAVEL', ascending=True)

drones_sumiram_com_email = drones_sumiram_com_email[['CODIGO_AERONAVE', 'DATA_VALIDADE_2023', 'RESPONSAVEL']]
drones_sumiram_com_email

email_df = email_df[['CODIGO_AERONAVE', 'DATA_VALIDADE_sisant', 'RESPONSAVEL']]

# Merging everything and taking of duplicates
drones_email_total = pd.merge(email_df, drones_sumiram_com_email, on=['CODIGO_AERONAVE', 'RESPONSAVEL'], how='left')

drones_email_total.drop(columns={'DATA_VALIDADE_2023'}, inplace=True)

drones_email_total['RESPONSAVEL'].fillna('nao contem email', inplace=True)

# Applying vencido() function
# Defining function to see if the Drone if 'VENCIDO' or 'NAO VENCIDO'
def vencido_new(df):
    if df['DATA_VALIDADE_sisant'] < pd.Timestamp('today'):
        return 'VENCIDO'
    else:
        return 'NAO VENCIDO'
drones_email_total['VENCIDO'] = drones_email_total.apply(vencido_new, axis=1)

# Displaying and saving final result
display(drones_email_total)
drones_email_total.to_csv('drones_vencidos_completo.csv')
```
